# DECISION TREES

### What is it?
This is a supervised machine learning algorithm which provides a visual representation and clear logic path for decision making for the data provided.

### When to use it?
Decision trees can be used when the target is known categories. They should be used when you want detailed information about how to traverse the variables to get to different target values. Unlike SVM and Neural Networks, Decision Trees are very transparent in the logic it uses to make decisions.


### Limitations?
1. Even small changes in the input can sometimes cause large changes in the output.
2. Seemly small changes in the data can cause major changes and might even be cause for redrawing the tree. These could be changes like changing variables, excluding duplication information, or altering the sequence midway.
3. Decision trees are built on expectations. Bad expectations or omitting a causal factor can lead to bad decision trees.
4. Large Decision trees can be very complex and time consuming to create.
5. Computing probabilities for each branch, determining best split for each node, can be complicated and require much prerequisite knowledge.
6. Decision trees look at only one variable at a time which is sometimes not true to reality.

For definitions of machine learning terms visit: https://developers.google.com/machine-learning/glossary/

```python
from sklearn.tree import DecisionTreeClassifier, export_graphviz
from sklearn import tree
from IPython.display import SVG
from graphviz import Source
from IPython.display import display
from ipywidgets import interactive
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import datasets
```

# Breast Cancer Data

```python
# load dataset
data = datasets.load_breast_cancer()
data
```

```python
# feature matrix
X = data.data


# target vector
y = data.target

y
```

```python
# class labels
labels = data.feature_names
labels
```

```python
# print dataset description
print(data.DESCR)
estimator = DecisionTreeClassifier()
estimator.fit(X, y)

graph = Source(tree.export_graphviz(estimator, out_file=None
   , feature_names=labels, class_names=['0', '1', '2'] 
   , filled = True))
display(SVG(graph.pipe(format='svg')))
```

```python
# class labels
labels = data.feature_names
def plot_tree(crit, split, depth, min_split, min_leaf=0.2):
    estimator = DecisionTreeClassifier(random_state = 0 
      , criterion = crit
      , splitter = split
      , max_depth = depth
      , min_samples_split=min_split
      , min_samples_leaf=min_leaf)
    estimator.fit(X, y)
    graph = Source(tree.export_graphviz(estimator
      , out_file=None
      , feature_names=labels
      , class_names=['0', '1', '2']
      , filled = True))
   
    display(SVG(graph.pipe(format='svg')))
    return estimator
inter=interactive(plot_tree 
   , crit = ["gini", "entropy"]
   , split = ["best", "random"]
   , depth=[1,2,3,4]
   , min_split=(0.1,1)
   , min_leaf=(0.1,0.5))
display(inter)
```
